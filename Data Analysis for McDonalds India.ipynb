```python
#Lib initialization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv('McDonalds_India_Menu_Feature.csv')
display(df)
```

```python
#Checking null data
df.isnull().sum()
```

```python
#Checking data descriptive
df.describe()
```

```python
# Let see the menu variety from Menu Category column
print(df['Menu Category'].unique())
```

From the output above, which menu category that has most items in it?

```python
menuitems = df['Menu Category'].value_counts()
display(menuitems)

menulabel = ['McCafe Menu','Regular Menu','Beverages Menu','Breakfast Menu','Gourmet Menu','Condiments Menu','Desserts Menu']

data = px.data.tips()
fig = px.pie(data, 
             values=menuitems, 
             names=menulabel, 
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

From the output, we can conclude that the McCafe Menu Category has the most variety items in it with 36.2% of 100%. Therefore, we can observe another aspect of McDonald's India from the nutrition menu side.

What kind of menu category that has the biggest Energy per kCal?

```python
# Here, I am going to get some values from the table and I sorted the mean value from Energy per kCal based on the 'Menu Category' column
df[['Menu Category', 'Energy (kCal)']].groupby('Menu Category').mean().sort_values(by = 'Energy (kCal)', ascending = False)
```

```python
# I am going to make a boxplot to visualize the data
plt.figure(figsize = (12, 6))
sns.set_style('darkgrid')
ax = sns.boxplot(x='Menu Category',
                 y='Energy (kCal)',
                 data=df, 
                 palette = 'dark')
plt.setp(ax.artists, 
         alpha=0.7, 
         linewidth=2)
plt.xticks(rotation=45)
plt.title('Energy Boxplot Visualization')
```

From this output, we can see that Gourmet menu category is the most menu category that has energy while the condiments menu has the least energy in it.

What about the protein?

```python
display(df[['Menu Category', 'Protein (g)']].groupby('Menu Category').mean().sort_values(by = 'Protein (g)', ascending = False))

plt.figure(figsize = (12, 6))
sns.set_style('darkgrid')
ax = sns.boxplot(x='Menu Category', 
                 y='Protein (g)', 
                 data=df, 
                 palette = 'dark')
plt.setp(ax.artists, 
         alpha=0.7, 
         linewidth=2)
plt.xticks(rotation=45)
plt.title('Protein Boxplot Visualization')
```

This output tells us that Gourmet Menu category still the number one which has a lot protein in it, but in this case Beverages Menu category has the least Protein while Condiments has more Protein than Beverages with a mean difference of 0.462876 grams

Hmm interesting... what about the Total Fat? Any differences from the two above? Let us see

```python
display(df[['Menu Category', 'Total fat (g)']].groupby('Menu Category').mean().sort_values(by = 'Total fat (g)', ascending = False))

plt.figure(figsize = (12, 6))
sns.set_style('darkgrid')
ax = sns.boxplot(x='Menu Category',
                 y='Total fat (g)', 
                 data=df, 
                 palette = 'dark')
plt.setp(ax.artists, 
         alpha=0.7, 
         linewidth=2)
plt.xticks(rotation=45)
plt.title('Total Fat Boxplot Visualization')
```

In this case, we can conclude that Gourmet Menu category still has the most total fat in it, while beverages, again has the least total fat from the other menu category.

We should try another parameter, which we can observe from the mean of Total Carbohydrate.

```python
display(df[['Menu Category', 'Sodium (mg)']].groupby('Menu Category').mean().sort_values(by = 'Sodium (mg)', ascending = False))

plt.figure(figsize = (12, 6))
sns.set_style('darkgrid')
ax = sns.boxplot(x='Menu Category',
                 y='Sodium (mg)', 
                 data=df, 
                 palette = 'dark')
plt.setp(ax.artists, 
         alpha=0.7, 
         linewidth=2)
plt.xticks(rotation=45)
plt.title('Sodium Boxplot Visualization')
```

The standings of the output still almost the same as before, but we can see that there is changes in the positions between index 2 until index 5.

We are not finished yet, we need to analyze again with Total Sugar (g) as parameter.

```python
display(df[['Menu Category', 'Total Sugars (g)']].groupby('Menu Category').mean().sort_values(by = 'Total Sugars (g)', ascending = False))

plt.figure(figsize = (12, 6))
sns.set_style('darkgrid')
ax = sns.boxplot(x='Menu Category',
                 y='Total Sugars (g)',
                 data=df, palette = 'dark')
plt.setp(ax.artists,
         alpha=0.7, 
         linewidth=2)
plt.xticks(rotation=45)
plt.title('Total Sugars Boxplot Visualization')
```

From couple of outputs that we already generate before, we can conclude that the Gourmet Menu category contains the most nutrition in a serving situation on average. Meanwhile, it contains more macroproteins such as: protein, fat and it has less micronutrients such as: total sugars contents.

Now, the question is is there any correlation between those parameters?

```python
# Dropping certain columns to make it clear

corr = df.drop(['Energy (kCal)mean','Protein (g)mean','Total fat (g)mean','Total carbohydrate (g)mean','Total Sugars (g)mean','Sodium (mg)mean','Per Serve Size_No units'], axis = 1).corr()
plt.figure(figsize = (12,10))
display(corr.head())

sns.heatmap(corr[(corr >= 0.5)|(corr <= -0.4)], 
            cmap = 'gist_earth', 
            vmax = 1, 
            vmin = -1,
            linewidths = 0.1,
            annot = True,
            annot_kws = {'size':8},
            square=True);
```

From the output above, we can assume that we have a group of ranges that can be used as reference to labelling:

*   0.8 - 1 = Strong correlation between 2 parameters
*   0.6 - 0.79 = Good correlation between 2 parameters
*   0.5 - 0.59 = Okay correlation between 2 parameters
*   Below 0.5 = Weak/no correlation between 2 parameters

Okay, this is the correlation table that we already provided. We need to specify it to see more advance. I will take the strongest correlation which has the value close to one, I took 2 strongest samples:

*   Energy - Total Fat
*   Total Sugar - Added Sugar


```python
# Providing the correlation based on the strongest value

fig, ax = plt.subplots(1,2)

fig.set_figheight(15)
fig.set_figwidth(20)

sctplt1 = sns.scatterplot(x = 'Energy (kCal)',
                          y = 'Total fat (g)',
                          ax = ax[0],
                          color = 'green',
                          data = df)

sctplt2 = sns.scatterplot(x = 'Total Sugars (g)',
                          y = 'Added Sugars (g)',
                          ax = ax[1],
                          color = 'black',
                          data = df)

sctplt1.set_title('Correlation plot between Energy and Total Fat')
sctplt2.set_title('Correlation plot between Total Sugars and Added Sugars ')

plt.show()
```

From the output above, we can conclude that if menu category contains high value of energy, we can see that the menu has high value of total fat too, which is from the "green dots" scatter plot, while the connection between added sugars and total sugars is similar.

As we already have the 2 of the strongest correlation, I choose one of those two to deeper than before. 

```python
plt.figure(figsize=(25,10),
           dpi = 200)

corrcheck = df.drop(['Energy (kCal)mean',
                     'Protein (g)mean',
                     'Total fat (g)mean',
                     'Total carbohydrate (g)mean',
                     'Total Sugars (g)mean',
                     'Sodium (mg)mean',
                     'Per Serve Size_No units'],
                     axis = 1)

corrcheck.corr()['Energy (kCal)'].sort_values(ascending = False).plot(kind ='bar',
                                                               color = 'green')
```

Furthermore, through this chart we can summrize that the correlation between the energy and the first feature (which is "Energy") is higher compared to the other features (e.g.: Total fat, Sodium, Protein, etc.).

Now, we do have knowledge about the analysis of nutritients content that is on the menu category and we know about correlation between these parameters too, which we can dig deeper to analyzed about the items that included in menu category, e.g.: Spicy Peneer Wrap, American Veg Burger, McAloo Tikki Burger, etc.

And now, what if we try to analyze the items that has least and most nutritient contents from the data frame. The first one is which menu item that has least and highest value of energy (kCal)?

```python
# Minimum Energy (kCal)
display(df[df['Energy (kCal)'] == df['Energy (kCal)'].min()])

# Maximum Energy (kCal)
display(df[df['Energy (kCal)'] == df['Energy (kCal)'].max()])
```

From the output above, we can conclude that the "Vedica Natural Mineral Water" has the least energy which has the value of 0 (probably it just a mineral water), while the "Chicken Cheese Lava Burger" has the highest energy which is over 800 (kCal) per serving.

Now, what about the total protein that contains in these menu items? Let us see.

```python
# Minimum Protein
display(df[df['Protein (g)'] == df['Protein (g)'].min()])

# Maximum Protein
display(df[df['Protein (g)'] == df['Protein (g)'].max()])
```

As we can see, there is a couple of items that does not have protein in their contents which is dominated by drinks food category, while "Chunky Chipotle American Burger Chicken" has the highest with a value of 39.47 grams.

Now, we want to see the number one standings in total fat contents.

```python
# Minimum Total Fat
display(df[df['Total fat (g)'] == df['Total fat (g)'].min()])

# Maximum Total Fat
display(df[df['Total fat (g)'] == df['Total fat (g)'].max()])
```

From the output that we already generated, we can see that again drinks dominated for the least total fat that does not have any fat in them, while "Chicken Cheese Lava Burger" has the most fat in it.

Now let see the value of the cholestrol.


```python
# Minimum Cholesterol
display(df[df['Cholesterols (mg)'] == df['Cholesterols (mg)'].min()])

# Maximum Cholesterol
display(df[df['Cholesterols (mg)'] == df['Cholesterols (mg)'].max()])
```

We can summarize that "McSpicy Premium Chicken Burger" has the highest value of cholestrols while the least, again dominated by drinks.

Now, we move on to the next thing that we need to analyze, which is total sugars.

```python
# Minimum Total Sugars (g)
display(df[df['Total Sugars (g)'] == df['Total Sugars (g)'].min()])

# Maximum Total Sugars (g)
display(df[df['Total Sugars (g)'] == df['Total Sugars (g)'].max()])
```

As we can see now the output now dominated by drinks for the highest and the least total sugars that contain in the menu items. "L1 Coffee", "Coke Zero Can", "Vedica Natural Mineral Water" does not contain any sugars in it while "Large Fanta Orange" has the highest value of sugars in it with the value of 64.22 grams per 544 ml.

Now, the last one. The sodium.


```python
# Minimum Total Sodium (mg)
display(df[df['Sodium (mg)'] == df['Sodium (mg)'].min()])

# Maximum Total Sodium (mg)
display(df[df['Sodium (mg)'] == df['Sodium (mg)'].max()])
```

After we generated previous output, we can see that "L1 Coffee" and "Piri Piri McSpicy Chicken Burger" does not have any sodium in it, while "Ghee Rice with McSpice Fried Chicken" has the highest value of sodium in it with 2399.49 mg per serving.
